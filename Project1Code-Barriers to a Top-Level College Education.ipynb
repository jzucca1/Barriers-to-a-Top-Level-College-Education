```python
#plot of starting salaries vs mid-career median salaries
from scipy import stats
x1 = starting['Starting Median Salary']/1000
y1 = starting['Mid-Career Median Salary']/1000
slope, intercept, r_value, p_value, std_err = stats.linregress(x1,y1)
line = slope*x1+intercept
pls.plot(x1,y1,'o', x1, line)
pls.title("Starting vs Mid-Career Median Salaries")
pls.xlabel("Starting Salary (in $1000s)")
pls.ylabel("Mid-Career Mediain Salary (in $1000s)")
```
